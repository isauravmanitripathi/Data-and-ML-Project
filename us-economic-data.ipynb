```python
import pandas as pd

# Load the cleaned dataset
data = pd.read_csv("datasets/cleaned_us_gov_data.csv")

# Step 1: Remove rows with placeholder values ('....*.') in key columns selectively
# Define the quarterly columns
quarterly_columns = ['2021 Q1', '2021 Q2', '2021 Q3', '2021 Q4', '2022 Q1', '2022 Q2', '2022 Q3', '2022 Q4', '2023 Q1', '2023 Q2']

# Remove rows where any of the quarterly columns contain placeholder values like '....*.'
data = data[~data[quarterly_columns].apply(lambda x: x.astype(str).str.contains(r'\.\.\.\.\*\.', na=False)).any(axis=1)]

# Step 2: Retain numeric descriptions but focus on rows containing actual labels
# Convert Description column to string to prevent errors and identify rows with meaningful labels
# Example labels to retain could include "Gross domestic product (GDP)"
meaningful_rows = data['Description'].astype(str).str.contains("Gross domestic product|expenditures|investment|income", case=False, na=False)

# Retain rows that either match the meaningful labels or are numeric
data_filtered = data[meaningful_rows | data['Description'].str.isnumeric()]

# Step 3: Convert quarterly columns to numeric, handling non-numeric entries gracefully
for col in quarterly_columns:
    data_filtered[col] = pd.to_numeric(data_filtered[col], errors='coerce')

# Display the cleaned dataset structure and sample rows
print("Cleaned Data Structure:")
print(data_filtered.info())
print("\nSample Data:")
print(data_filtered.head())
```

```python
import pandas as pd

# Load the dataset
data = pd.read_csv("datasets/cleaned_us_gov_data.csv")

# Step 1: Replace placeholder values ('....*.') in all columns with NaN
data.replace("....*.", pd.NA, inplace=True, regex=True)

# Step 2: Convert quarterly columns to numeric (this will set any remaining non-numeric values to NaN)
quarterly_columns = ['2021 Q1', '2021 Q2', '2021 Q3', '2021 Q4', '2022 Q1', '2022 Q2', '2022 Q3', '2022 Q4', '2023 Q1', '2023 Q2']
for col in quarterly_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Step 3: Save the cleaned data to a new CSV file
data.to_csv("datasets/final_cleaned_us_gov_data.csv", index=False)

# Confirm the operation
print("Placeholder values replaced with NaN, and file saved as 'final_cleaned_us_gov_data.csv'")
```

```python
import matplotlib.pyplot as plt

# Filter the data for "Gross domestic product (GDP)"
gdp_data = data_v3[data_v3['2021 Q1'] == 'Gross domestic product (GDP)']

# Select quarterly columns and transpose them for plotting
quarters = ['2021 Q2', '2021 Q3', '2021 Q4', '2022 Q1', '2022 Q2', '2022 Q3', '2022 Q4', '2023 Q1', '2023 Q2']
gdp_values = gdp_data[quarters].values.flatten().astype(float)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(quarters, gdp_values, marker='o', color='b', linestyle='-', linewidth=2, markersize=6)

# Adding title and labels
plt.title('Gross Domestic Product (GDP) Over Time', fontsize=16)
plt.xlabel('Quarter', fontsize=12)
plt.ylabel('GDP Value', fontsize=12)
plt.xticks(rotation=45)
plt.grid(visible=True, linestyle='--', linewidth=0.5)

# Show plot
plt.show()
```
